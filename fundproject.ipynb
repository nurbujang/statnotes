## 2.3 Plots

### 2.3.1 Violin Plot for Numerical Data

While a histogram (Wikipedia Contributors, 2019c) with Kernel Density Estimation (KDE) (Wikipedia Contributors, 2019f) is an appropriate plot for each numerical variable, it lumps the observation for all three Iris species together, making it less useful for data interpretation. Therefore, the most appropriate plot which also groups Iris species separately for each flower attribute and provides the most statistical information and data distribution is the Violin Plot.

The Violin plot is hybrid of a box plot and a kernel density plot and it compares probability distributions between two variables (Wikipedia, 2020c, Seaborn.pydata.org violinplot, n.d.). The Violin curve's width indicates frequency in each section and is computed using KDE (Yi, 2021). While the box plot only outlines the summary statistics (Carron, 2021), the Violin Plot shows both the summary statistics and each variable's density, thus offering a more detailed representation of the shape of the data.

Violin Plots with clearer overlapping boxplots were generated for all numerical columns according to C (2021). To optimise spacing between subplots, fig.tight_layout was used according to GeeksforGeeks spacing (2020). Spacing between subplots and title (y=1.01) were adjusted according to Ceylan (2018).

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

fig, axs = plt.subplots(1, len(columns)-1, figsize=(20,6)) #
for i in range(0,len(columns)-1): 
    ax=sns.violinplot(x='class', y=df[columns[i]], data=df, ax=axs[i])
    sns.boxplot(x='class', y=df[columns[i]], data=df, palette='turbo', width=0.12,
            boxprops={'zorder': 2}, ax=ax)
    axs[i].set_ylabel(columns[i])

plt.suptitle('Violin Plot for Sepal and Petal Attributes of Three Iris species', color='#3D3F38', 
             fontweight='bold', size=15) 
fig.tight_layout()
plt.show();
```

The Violin Plot shows that in general, petal length and width for *I. setosa* are identifiably much smaller than *I.versicolor* and *I.virginica*, but it is not significantly different in terms of sepal attributes. *I. versicolor* and *I. virginica* are not significantly different in all petal and sepal attributes. 

For sepal length, *I. setosa* has a higher density of data around the median and is symmetrically distributed. Observations in *I. versicolor* and virginica are more wide-spread and right-skewed.

For sepal width, all three species are not significantly different from one another with higher accumulation of data around the median. *I. setosa* has long tails on both lower and upper ends, suggesting a wide range. *I. versicolor* is negatively-skewed and there are outliers present on the lower and upper ends of *I. virginica*.

For petal length, while it is symmetrically distributed with a big cluster of data around the median, outliers are also present on both lower and upper ends of *I. setosa*. *I. versicolor* and *I. virginica* data are more dispersed but slightly left-skewed.

Interestingly, for petal_width in *I. setosa*, the min value is the same as the first quartile, meaning that the data distribution is relatively concentrated above this minimum value and right-skewed. Outliers are also present above the upper end. Data is more spread out in *I. versicolor* and *I. virginica* and both are also positively skewed.
