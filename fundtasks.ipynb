## Task 1 : Collatz Conjecture
> The Collatz conjecture is a famous unsolved problem in mathematics. The problem is to prove that if you start with any positive integer x and  repeatedly apply the function f ( x ) below, you al ways get stuck in the repeating sequence 1, 4, 2, 1, 4, 2, . . .

\begin{align*}
f(x) = \begin{cases} 
    x \div 2 & \text{if } x \text{ is even} \\
    3x + 1 & \text{if } x \text{ is odd}
\end{cases}
\end{align*}

> For example, starting with the value 10, which is an even number, we divide it by 2 to get 5. Then 5 is an odd number so, we multiply by 3 and add 1 to get 16. Then we repeatedly divide by 2 to get 8, 4, 2, 1. **Once we are at 1, we go back to 4 and get stuck in the repeating sequence 4, 2, 1 as we suspected.**

### Task Description:
> The task is to verify, using Python, that the Collatz conjecture is true for the first 10000 positive integers.


### Method and Implementation

According to the Collatz Conjecture, all numbers will end with 1 before going into another loop of 4,2,1 (Honner, 2020, Stack Overflow, n.d., Sao, 2021). Hence, the strategy is to stop the While loop at 1 (w3schools, n.d.).

Python Functions (W3 Schools, 2019) and While loop (w3schools, n.d.) were used for the conditions on odd and even numbers. Python list append method was used to keep adding on to the list until it reaches 1 (www.w3schools.com, n.d.).

```python
def collatz(x):
    
    clist = [x] # numbers will be in a list

    while x != 1:
        if x % 2 == 0:
            x = x // 2 # If x is even, divide it by two
        else:
            x = (3 * x) + 1 # if x is odd, multiply by 3, then add 1
        clist.append(x) # list will append

    return clist
```

Python For Loop For i in Range was used (Thompson, 2021) to determine the limit, with if, else conditions (W3Schools, 2019, GeeksforGeeks, 2017) to verify that the last number in the list (Stack Overflow, n.d.) is 1.

Lastly, the verify_collatz function was called and printed out the verification/nonverification statement.

```python
# for verification of the first 10000 positive integers

def verify_collatz(limit):
    for i in range(1, limit + 1):
        clist = collatz(i)

# if NOT verified, output will NOT end with 1
    if clist[-1:] != [1]: # slicing and comparing
        print(f"The Collatz Conjecture is not true for x = {i}")
        return
    else:
        print("The Collatz Conjecture is true the first", limit, "positive integers.")

verify_collatz(10000) # call the function 
```

### Conclusion

The Collatz Conjecture is verified to be true for the first 10000 positive integers.

```python
for s in string_columns:
    print(s, df[s].astype('category').cat.categories) # print categories in string columns

for n in numeric_columns:
    try:
        df[n].astype(int) # run a code that might cause exceptions
    except:
        print(f'column {n} cannot be converted to int.') # run this code when there is an exception
        print(f'does {n} have a null value?', df[n].isna().any()) # does this column contain a missing value?
    else:
        print(f'column {n} can be converted to int') # run this code if there are no exceptions
```

The flipper length and body mass are considered floats because they contain null values (pandas.pydata.org isnull, n.d.) which cannot be converted into int values. df.info (w3resource, 2022b) would also give the same information in terms of null values. ‌

## **Task 4 : Head Probability** 
> Suppose you are flipping two coins, each with a probability p of giving heads.

### Task Description:
> The task is to plot the entropy of the total number of heads versus p.

### Methods and Implementation

First, numpy for arrays and random numbers (numpy.org numpy.array, n.d.), matplotlib for plots (matplotlib.org pyplot , n.d.) and scipy stats entropy function to calculate entropy values (docs.scipy.org entropy, n.d.) were imported.

```python
import numpy as np # for numerical arrays and random numbers
import matplotlib.pyplot as plt  # for plots
from scipy.stats import entropy
```

Random binomial (Numpy.org random.binomial , 2021) was used because the probability is only either head or tail.

```python
# USE RANDOM BINOMIAL (n, p, size=None), because you can only get head or tail

two_rand_coins = np.random.binomial(2, 0.5, 1000) # flip 2 fair coins, probability of getting heads is 0.5, flip 100 times
two_rand_coins # show array
```

Then, a plot was created to visualize the probabilities according to Matplotlib.org pyplot (n.d.).

```python
# Create an empty plot.
fig, ax = plt.subplots(figsize=(12, 4))

# Count the unqiue values in the array.
x, counts = np.unique(two_rand_coins, return_counts=True)

# Create a bar chart of rand_coins.
ax.bar(x, counts)

# Make the x axis look sensible.
ax.set_xticks(x);

# Output will show MORE the chance of getting 1 head than 0 heads or 2 heads at 100 flips.
# It will even out if we increase number of flips
```

A function to calculate the entropy was defined using def (Chris, 2022), and the return statement (GeeksforGeeks, 2019) was used to end the execution of the function calculate_entropy with values probabilities (getting head only, head and tail, and tail only) and base 2.

```python
# Function to calculate entropy
def calculate_entropy(p):
    t = 1 - p  # probability of getting tails
    probabilities = [p**2, 2 * p * t, t**2]  # getting head only, head and tail, and tail only
    return entropy(probabilities, base=2) # base 2 because it's either head or tail
```

Then, the values for p using numpy linspace to return 500 evenly spaced numbers between 0 and 1 were generated (numpy.org linspace, n.d.).

```python
# Generate values for p
total_heads = np.linspace(0, 1, 500) # get 500 evenly-spaced numbers between 0 and 1

total_heads
```

Following that, the entropy was calculated for each p value (docs.scipy.org entropy, n.d.).

```python
# Calculate entropy for each value of p
entropy_values = [calculate_entropy(h) for h in total_heads]

entropy_values
```

Finally, a plot for entropy of the total number of heads VS probability of getting heads was generated (matplotlib.org pyplot , n.d.).

```python
# Plot the entropy
plt.plot(total_heads, entropy_values, label='Entropy')
plt.title('Entropy of the total number of heads vs. probability of heads for two coins')
plt.xlabel('Probability of heads')
plt.ylabel('Entropy')
plt.show()
```

### Methods and Implementation

First, pandas (data loading), seaborn (statistical graphics), matplotlib pyplot (plots) and os (interaction with operating system) libraries were imported. Warning controls were issued to suppress warnings in the code (Docs.python.org, n.d., Deepak, 2023).

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

The famous penguins data set (Waskom, 2022, Pandey, 2020) was used. After it was imported using os.path.join according to Pandas pydata read_csv (n.d.) and GeeksforGeeks (2019), df.head was executed to show a few lines of the data frame.

```python
file_path = os.path.join('~/Desktop/ATU/W23/FUNDAW2023/fundaw23', 'data', 'penguins.csv')

df = pd.read_csv(file_path)
df.head()  # default is 5
```

First, the style  of the plot was set to have tick marks on the axes. 

Then, For loop (W3Schools, 2019b) was used to go through each column of the df. Python conditions if, else (W3Schools, 2019a) was used to differentiate between numerical and categorical datatypes. Pandas dtype was used to specify the datatypes of each column.

Violin Plots were generated for numerical columns using seaborn violin plot (Waskom, n.d.) according to Keshav (2021) and Kaggle.com (2022), while seaborn countplot was generated for categorical data with species as hue (Waskom, 2012). 

```python
# set the style for the plots
sns.set(style="ticks")

for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        plt.figure(figsize=(8, 6))
        sns.violinplot(x='species', y=column, data=df, box_visible=True, palette='pastel')
        plt.title(f'Violin Plot for {column}')
        plt.show();

    else:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=column, data=df, hue='species', palette='pastel', edgecolor="black")
        plt.title(f'Count Plot for {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.legend(loc=1)
        plt.show();
```

### Conclusion

Violin Plots for the numerical variables and Count Plots for categorical variables in the penguin data set were generated.

Results show that the species count in the dataset in descending order is Adelie, Gentoo and Chinstrap.

Only Adelie was found in Togerson Island, only Adelie and Gentoo were found in Biscoe Island and Dream Island has Adelie and Chinstrap species only.

For bill length, data is more clustered around the median for Adelie with a symmetrical distribution, but more spread and left-skewed in Chinstrap and spread out and slightly right-skewed in Gentoo.

For bill depth, data is more spread out in Adelie, but more clustered around the median for Chinstrap and Gentoo.

For flipper length, data is symmetrically distributed with more concentrated around the median with outliers on both lower and upper ends for Adelie and Chinstrap. Gentoo is slightly right-skewed with most of the data concentrated around the median.

For body mass, most of the data are accumulated around the median in Adelie and Chinstrap, more so in the latter. Data is spread out and right-skewed in Gentoo.

The number of male and female are almost equal in all species in the dataset, except Gentoo with slightly more males than females.
