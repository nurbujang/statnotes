## Probability Density Function

$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $

https://en.wikipedia.org/wiki/Normal_distribution

![normalprobability](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/1024px-Normal_Distribution_PDF.svg.png)
![normalcumulative](https://upload.wikimedia.org/wikipedia/commons/thumb/c/ca/Normal_Distribution_CDF.svg/1024px-Normal_Distribution_CDF.svg.png)

The parameter  {\textstyle \mu } is the mean or expectation of the distribution (and also its median and mode), while the parameter  2 {\textstyle \sigma ^{2}} is the variance. The standard deviation of the distribution is  {\textstyle \sigma } (sigma). A random variable with a Gaussian distribution is said to be normally distributed, and is called a normal deviate. 

```python
np.e
```

```python
def f(x, mu=0.0, sigma=1.0): # 1 is nice easy value for sigma which cant be 0 or negative value, mu can be 0
  # The normalisation constant.
  norm = 1.0 / np.sqrt(2.0 * np.pi * sigma**2)
  
  # f(x)
  return norm * np.exp(-(x - mu)**2 / (2.0 * sigma**2))
```

```python
# Test the function. check red line in graph above
# the red line's y-axis is almost 0.4
f(0.0)
# very very close to 0.4 (very close to graph)
```

```python
# Create an x range.
x = np.linspace(-5, 5, 1001) 
# starting point, end point, 
# number of points u want between those 2 numbers (odd numbers will get symmetry)

# Get y values.
y = f(x)

# Show.
x, y
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Plot the function.
ax.plot(x, y);

# will get like the red line from graph
# peak is at y is almost 0.4
```

Reference: [6.4: Normal Approximation to the Binomial Distribution](https://stats.libretexts.org/Courses/Las_Positas_College/Math_40%3A_Statistics_and_Probability/06%3A_Continuous_Random_Variables_and_the_Normal_Distribution/6.04%3A_Normal_Approximation_to_the_Binomial_Distribution)

```python
# To superimpose f(x) on top of coin flipping plot

# Number of coins to flip.
n = 100

# Number of times to do it.
k = 10000 * n

# Probability of heads.
p = 0.5

# Flip two coins several times.
nflips = np.random.binomial(n, p, k)

# Count the number of times each value occurs.
uniques, counts = np.unique(nflips, return_counts=True)

# Create a blank plot.
fig, ax = plt.subplots(figsize=(12, 5))

# Create a bar chart.
ax.bar(uniques, counts / k) # if (uniques, counts), y scale will be 80000 at 100000 flips
# so count/k to get proportions. if count/k and f(x, mu=n*p), line is tall, bar is short
# so scale it down by y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p)))
# ax.bar(uniques, counts) # this y axis adjustment will have y scale to 80000

# Create an x range.
x = np.linspace(25, 75, 1001) # if (-5, 5, 1001) centerpoint will be at 50
#x = np.linspace(-5, 5, 1001)


# Get y values.
y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p))) # supply the right mu and sigma values
# miu is centerpoint (50)
# n=number of coins flipped, p=probability of head, 1-p= probability of tail
# we have to supply miu and sigma
# and if x = np.linspace(-5, 5, 1001), and
# if y=f(x) only, centerpoint will be moved to right
# if y = f(x, mu=n*p), centerpoint will be at 50
# y=f(x)
#y = f(x, mu=n*p)
# think: tail will never hit x axis, because you can never get <1 head
# sigma is bernoulli trials: number of coins x number of head x number of tails
# 100 x 0.5 x 0.5 = 25, square root 25 is 5, so std is 5
# 5 is some measure of how quickly this red line gets close to x axis
# rule of thumb: if u go 3 std away from mean in either direction, (mean=50, so 55 is std, 60 is 2 std etc)
# then the vast majority of area under the red curve is going to be under those 2 limits (35 and 65) 
# so when u flip 100 coins, it is very unlikely that u see number of heads <35 and >65


# Plot the function.
ax.plot(x, y, 'r-') # redline

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title(f'Flipping {n} coins {k} times')

# Add axis label.
ax.set_xlabel('No heads') # here we are interested in heads only
ax.set_ylabel('Proportion');
```
